Drawing Shapes OPENCV

```python
def convert_BGR_to_RGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

def plot_BGR_to_RGB(image,plt):
    plt.imshow(convert_BGR_to_RGB(image))
```

```python
board =  np.zeros(
                  (300,400,3) #shape of the image with 3 channels for RGB
                  , dtype="uint8" #type of the data
                 )

plot_BGR_to_RGB(board,plt)
plt.axis("off")
```

```python
(r,g,b) = image[0,0]
(r,g,b)
```

```python
green=(0,255,0)
red = (255,0,0)
blue=(0,0,255)
cv2.line(board,(0,0),(200,400),green)
cv2.line(board,(0,400),(400,0),red,5) #first param - image, start and end points, color and thickness (-1 is shape fill)
cv2.rectangle(board, (20,10), (100,100), green, 5) # Draw Rectange
cv2.rectangle(board, (400,200), (300,300), blue, -1)
(centerX,centerY) = (board.shape[1] // 2, board.shape[0] // 2) # Define the center point
cv2.circle(board,(centerX,centerY),50, red, 15 )# Draw Circle
plot_BGR_to_RGB(board,plt)
```
